```python
import numpy as np
import pandas as pd
```

```python
name=["mr ram", "miss rina", "master golu", "mrs. piku", "mr. lala", "miss billo", "master groot", "mrs. shila", "mr. nikku", 
      "master biba"]
age=[23, 36, 34, 22, 27, 22, 24 ,25, 26, 30]
qual= ["MSc", "BSc", "BCA", "MCA", "B.Tech","M.tech","BSC","MSC","B.Tech","MCA"]
en_mark= [56, 78, 89, 34, 56, 43, 56, 78, 89, 70]
sc_mark= [98, 65, 43, 32, 45, 54, 66, 82, 91, 42]
cs_marks=[65, 87, 43, 98, 65, 56, 78, 33, 39, 88]
data= pd.DataFrame({"Name":name, "Age":age,
                    "Qualification":qual,
                    "English marks": en_mark,
                    "science marks": sc_mark,
                    "computer marks": cs_marks},
                   index= np.arange(1,11))
data
```

```python
#name and age of student with max age
mn= data.Age.idxmax()
data.loc[2, ['Name','Age']]
```

```python
#number of students with masters degree
data.Qualification.str.startswith('M').sum()
```

```python
#qualification of unmarried male
ml=data.Name.str.startswith('master')
data.loc[ml,['Qualification']]
```

```python
#remove salutation
def remove_sal(names):
    s_p= names.index(" ")
    return names[s_p+1:]
Names=data.Name.apply(remove_sal)
data.insert(1, "without salutation", Names)
data
```

```python
#total marks column in last
data["Total marks"]= data.iloc[:,[-1,-2,-3]].sum(axis=1)
data
```

```python
# inserting a gender column
def gender(name):
    s_p= name.index(" ")
    if name[:s_p]=="mr" or name[:s_p]=="master":
        gen="Male"
    else:
        gen= "Female"
    return gen
Gender= data.Name.apply(gender)
data.insert(2, "Gender", Gender)
data
```
